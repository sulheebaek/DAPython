```python
def fstats(x):
    return Series([x.min(), x.max(), x.mean()], index=['min', 'max', 'mean'])
```

```python
price.apply(fstats)
```

```python
price.apply(fstats, axis=1).head()
```

```python
def fargs(x):
    return Series([x.idxmin(), x.idxmax()], index=['min_index', 'max_index'])
```

```python
price.apply(fargs)
```

```python
price.apply(fargs, axis=1).head()
```
